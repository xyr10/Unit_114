```python
# simple example

def fact(num):
    if num == 0:
        return 1
    return num*fact(num-1)
```

```python
from functools import lru_cache

@lru_cache
def fib(num):
    if num < 2:                   # base case (our exit condition but also an absolute truth)
        return num
    print(num)
    return fib(num-1) + fib(num-2) # recursive case
 
fib(50)
```
