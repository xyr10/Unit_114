```python
class SinglyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None
            
    def __init__(self):
        self.head = None
        
    def append(self, value):
        new_node = self.__Node(value)
        if not self.head:                      # state 1: when the list is empty
            self.head = new_node
        else:
            current = self.head                # state 2: when there is only one node and ...
            while current.next:                # state 3: when there is more than one neode in our list.
                current = current.next
            current.next = new_node
        
    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % current.data
            current = current.next
            while current:
                out += ", %s" % current.data
                current = current.next
        out += "]"
        return out
    
    def insert(self, index, value):
        pass
    
    def remove(self, value):
        pass
    
    def index(self, value):
        # return first index of value or ValueError if not present
        pass
    
```
